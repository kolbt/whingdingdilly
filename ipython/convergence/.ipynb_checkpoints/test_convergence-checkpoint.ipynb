```python
def find(name):
    home = os.path.expanduser("~")
    for root, dirs, files in os.walk(home):
        if name in dirs:
            return os.path.join(root, name)
```

```python
# First let's find all of our data
whingPath = find('whingdingdilly')
ipyPath = whingPath + '/ipython'
rootPath = ipyPath + '/convergence'
dataPath = rootPath + '/txtFiles'

# Go to the correct parent directory
os.chdir(rootPath)

# Get files from each source directory
Files = os.listdir(dataPath)
nFiles = len(Files)
```

```python
# Functions to sort my data with
def getFromTxt(fname, first, last):
    """Takes a string, text before and after desired text, outs text between"""
    start = fname.index( first ) + len( first )
    end = fname.index( last, start )
    myTxt = fname[start:end]
    return float(myTxt)
        
def singleSort(arr):
    """Sort an array the slow (but certain) way, returns original indices in sorted order"""
    # Doing this for PeR, PeS, xS in this case
    cpy = np.copy(arr)
    ind = np.arange(0, len(arr))
    for i in xrange(len(cpy)):
        for j in xrange(len(cpy)):
            # Sort by first variable
            if cpy[i] > cpy[j] and i < j:
                # Swap copy array values
                cpy[i], cpy[j] = cpy[j], cpy[i]
                # Swap the corresponding indices
                ind[i], ind[j] = ind[j], ind[i]  
    return ind

def indSort(arr1, arr2):
    """Take sorted index array, use to sort array"""
    # arr1 is array to sort
    # arr2 is index array
    cpy = np.copy(arr1)
    for i in xrange(len(arr1)):
        arr1[i] = cpy[arr2[i]]
```

```python
# Let's sort our respective lists
nList = []
for i in xrange(nFiles):
    nList.append(getFromTxt(Files[i], "N", ".txt"))
# Now sort the array of txtFile names
indArr = singleSort(nList)
indSort(Files, indArr)
for i in xrange(nFiles): print(Files[i])
```

```python
# Read the data for each parameter study into a pandas dataframe
os.chdir(dataPath)
sims = []
for i in xrange(nFiles):
    df = pd.read_csv(Files[i], sep='\s+', header=0)
    sims.append(df)
        
# Go back to the source folder (save figures here)
os.chdir(rootPath)        

# If you want to see what your data looks like
# display(all_sims[0][0])

# Get the name of each of the headers in the dataframes
list(sims[0])
```

```python
# Make sure all data is chronilogical
def chkSort(array):
    """Make sure array is chronilogical"""
    for i in xrange(len(array)-2):
        if array[i] > array[i+1]:
            print("{} is not greater than {} for indices=({},{})").format(array[i+1], array[i], i, i+1)
            return False
    return True

# Check to see if timesteps are in order

for i in xrange(nFiles):
    myBool = chkSort(sims[i]['Timestep'])
    if myBool is False:
        print("{} is not chronilogically sorted!").format(Files[i])
        exit(1)
    else:
        print("{} sorted... ").format(Files[i])
```

```python
# Add a column to the all_sims dataframe for number of clusters (of threshold size):
for i in xrange(nFiles):
    # New column for number of clusters
    sims[i]['nClust'] = sims[i]['Dense_tot'] / sims[i]['MCS']
    # Ger rid of NaN in favor of 0
    sims[i].fillna(0, inplace=True)

# display(all_sims[0][0])
list(sims[0])
```

```python
# Function that will sort wrt one variable
def singleSort(arr):
    for i in xrange(len(arr)):
        for j in xrange(len(arr)):
            if arr[i] < arr[j] and i > j:
                arr[i], arr[j] = arr[j], arr[i]
                
# Function to get conversion from timesteps to Brownian time
def computeTauPerTstep(epsilon):
    # This is actually indpendent of runtime :)
#     sigma = 1.0
#     threeEtaPiSigma = 1.0
#     runFor = 200
#     tauBrown = 1.0
#     tauLJ = ((sigma**2) * threeEtaPiSigma) / epsilon
#     dt = 0.00001 * tauLJ
#     simLength = runFor * tauBrown
#     totTsteps = int(simLength / dt)
#     tstepPerTau = int(totTsteps / float(simLength))
    kBT = 1.0
    tstepPerTau = int(epsilon / (kBT * 0.00001))
    return tstepPerTau

def theoryDenom(xS, peS, peF):
    xS /= 100.0
    xF = 1.0 - xS
    return 4.0 * ((xS * peS) + (xF * peF))

def theory(xS, peS, peF):
    kappa = 4.05
    xS /= 100.0
    xF = 1.0 - xS
    return ((3.0 * (np.pi**2) * kappa) / (4.0 * ((xS * peS) + (xF * peF))))

def areaCircle(r):
    return np.pi * (r ** 2)

sigma = 1.0
def nToLBox(N):
    A_box = (N * areaCircle(sigma / 2.0)) / 0.6
    L_box = np.sqrt(A_box)
    return L_box

def convergenceRatio(PeF, N):
    lp = PeF / 3.0
    lbox = nToLBox(N)
    lplb = lp / lbox
    return lplb

# Make an additional frame that gives total number of particles, and simulation parameters
params = []
paramList = []
for i in xrange(nFiles):
    partAll = sims[i]['Gas_tot'][0]
    partA = sims[i]['Gas_A'][0]
    partB = sims[i]['Gas_B'][0]
    pa = getFromTxt(Files[i], "pa", "_pb")
    pb = getFromTxt(Files[i], "pb", "_xa")
    xa = getFromTxt(Files[i], "xa", "_ep")
    ep = getFromTxt(Files[i], "ep", "_N")
    N = getFromTxt(Files[i], "N", ".txt")
    lplb = round(convergenceRatio(pb, N), 1)
    converT = computeTauPerTstep(ep)
    mytheoryDenom = theoryDenom(xa, pa, pb)
    if pa != 0 or pb != 0: 
        mytheory = theory(xa, pa, pb)
    else:
        mytheory = 0.0

    paramList.append((partAll, partA, partB, pa, pb, xa, ep, N, lplb, converT, mytheoryDenom, mytheory))

# Put the data for this parameter sweep into it's own dataframe
params = pd.DataFrame(paramList, columns=['partAll',
                                              'partA',
                                              'partB',
                                              'peA',
                                              'peB',
                                              'xA',
                                              'eps',
                                              'N',
                                              'lplb',
                                              'brownTime',
                                              'theoryDenom',
                                              'theory']
             )
pd.set_option('display.max_rows', 2)
display(params)
display(params['partA'][nFiles - 1])
```

```python
# We should average past a specific time, not an index number...
ssStartTime = 2.0
ssStopTime = 7.5
ssStartInd = []
ssStopInd = []

# Figure out what we should use for steady-state
for i in xrange(nFiles):
    if sims[i]['Timestep'].iloc[-1] / params['brownTime'][i] >= ssStartTime:
        startFlag = 0
        for j in xrange(len(sims[i]['Timestep'])):
            if sims[i]['Timestep'][j] / params['brownTime'][i] >= ssStartTime and startFlag == 0:
                startFlag = 1
                ssStartInd.append(j)
            if sims[i]['Timestep'][j] / params['brownTime'][i] >= ssStopTime:
                ssStopInd.append(j-1)
                break
                
print(ssStartInd)
print(ssStopInd)
```

```python
# Make list of steady state column headers
headers = list(sims[0])
headers.remove('Timestep')
SS = pd.DataFrame(columns=headers)
stdDev = pd.DataFrame(columns=headers)
var = pd.DataFrame(columns=headers)
# Initialize dataframes
for j in xrange(nFiles):
    SS.loc[j] = [0] * len(headers)
    stdDev.loc[j] = [0] * len(headers)
    var.loc[j] = [0] * len(headers)

# Make dataframe of steady-state data
for j in xrange(nFiles):
    # Loop through each column (aside from tstep column)
    for k in range(1, len(sims[j].iloc[1])):
        # Compute mean of data after steady-state time (25tb) in jth column of ith file
        avg = np.mean(sims[j].iloc[ssStartInd[j]:ssStopInd[j], k])
        SS[headers[k-1]][j] = avg
        # Compute the standard deviation and variance in this data
        stdDevor = np.std(sims[j].iloc[ssStartInd[j]:ssStopInd[j], k])
        stdDev[headers[k-1]][j] = stdDevor
        var[headers[k-1]][j] = stdDevor ** 2

# Normalize by number of particles
for j in xrange(nFiles):
    if params['partA'][j] != 0:
        SS['Gas_A'][j] /= params['partA'][j]
        SS['Dense_A'][j] /= params['partA'][j]
        SS['Lc_numA'][j] /= params['partA'][j]
        # Now my standard error is a percentage
        stdDev['Gas_A'][j] /= params['partA'][j]
        stdDev['Dense_A'][j] /= params['partA'][j]
        stdDev['Lc_numA'][j] /= params['partA'][j]
        var['Gas_A'][j] /= params['partA'][j]
        var['Dense_A'][j] /= params['partA'][j]
        var['Lc_numA'][j] /= params['partA'][j]

    if params['partB'][j] != 0:
        SS['Gas_B'][j] /= params['partB'][j]
        SS['Dense_B'][j] /= params['partB'][j]
        SS['Lc_numB'][j] /= params['partB'][j]
        stdDev['Gas_B'][j] /= params['partB'][j]
        stdDev['Dense_B'][j] /= params['partB'][j]
        stdDev['Lc_numB'][j] /= params['partB'][j]
        var['Gas_B'][j] /= params['partB'][j]
        var['Dense_B'][j] /= params['partB'][j]
        var['Lc_numB'][j] /= params['partB'][j]

SS['Gas_tot'][:] /= params['partAll'][:]
SS['Dense_tot'][:] /= params['partAll'][:] 
SS['Lg_clust'][:] /= params['partAll'][:] 
SS['MCS'][:] /= params['partAll'][:]
stdDev['Gas_tot'][:] /= params['partAll'][:]
stdDev['Dense_tot'][:] /= params['partAll'][:] 
stdDev['Lg_clust'][:] /= params['partAll'][:] 
stdDev['MCS'][:] /= params['partAll'][:]
var['Gas_tot'][:] /= params['partAll'][:]
var['Dense_tot'][:] /= params['partAll'][:] 
var['Lg_clust'][:] /= params['partAll'][:] 
var['MCS'][:] /= params['partAll'][:]

SS['Gas_A'][:] *= 100.0
SS['Gas_B'][:] *= 100.0
SS['Gas_tot'][:] *= 100.0
SS['Dense_A'][:] *= 100.0
SS['Dense_B'][:] *= 100.0
SS['Dense_tot'][:] *= 100.0
SS['Lc_numA'][:] *= 100.0
SS['Lc_numB'][:] *= 100.0
SS['Lg_clust'][:] *= 100.0
SS['MCS'][:] *= 100.0
stdDev['Gas_A'][:] *= 100.0
stdDev['Gas_B'][:] *= 100.0
stdDev['Gas_tot'][:] *= 100.0
stdDev['Dense_A'][:] *= 100.0
stdDev['Dense_B'][:] *= 100.0
stdDev['Dense_tot'][:] *= 100.0
stdDev['Lc_numA'][:] *= 100.0
stdDev['Lc_numB'][:] *= 100.0
stdDev['Lg_clust'][:] *= 100.0
stdDev['MCS'][:] *= 100.0
var['Gas_A'][:] *= 100.0
var['Gas_B'][:] *= 100.0
var['Gas_tot'][:] *= 100.0
var['Dense_A'][:] *= 100.0
var['Dense_B'][:] *= 100.0
var['Dense_tot'][:] *= 100.0
var['Lc_numA'][:] *= 100.0
var['Lc_numB'][:] *= 100.0
var['Lg_clust'][:] *= 100.0
var['MCS'][:] *= 100.0
    
pd.set_option('display.max_rows', 6)
display(SS)
```
